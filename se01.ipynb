```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
    a = 0.2
    b = -2
    c = 20

    temperature = a * (time**2) + b * time + c
    return temperature, a, b, c

def main():
    time_values = np.linspace(0, 10, 100)

    temperature_hardcoded, a, b, c = quadratic_model(time_values)

    plt.plot(time_values, temperature_hardcoded, label='hard-coded coefficient')
    plt.xlabel('time')
    plt.ylabel('temperature')

    coefficients_text = f'a = {a}\nb = {b}\nc = {c}'
    plt.text(7, quadratic_model(7)[0], coefficients_text, bbox=dict(facecolor='white', alpha=0.5))

    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation (Hard-coded Coefficients)')

    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    # List of sets of coefficients
    coefficient_sets = [
        (1, 2, 3),
        (-1, 0, 1),
        (2, -1, 1),
    ]

    # Generate time values
    time_values = np.linspace(0, 10, 100)

    # Plot each set of coefficients
    for i, coefficients in enumerate(coefficient_sets):
        a, b, c = coefficients
        temperature_values = quadratic_model(time_values, a, b, c)
        label = f'Set {i + 1}: a={a}, b={b}, c={c}'
        plt.plot(time_values, temperature_values, label=label)

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation (Multiple Coefficient Sets)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    # Hardcoded coefficients
    a_hardcoded = 0.2
    b_hardcoded = -2
    c_hardcoded = 20

    # Get user input for coefficients
    a_user = float(input("Enter coefficient 'a': "))
    b_user = float(input("Enter coefficient 'b': "))
    c_user = float(input("Enter coefficient 'c': "))

    # Generate time values
    time_values = np.linspace(0, 10, 100)

    # Calculate temperature using hardcoded coefficients
    temperature_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)

    # Calculate temperature using user-input coefficients
    temperature_user = quadratic_model(time_values, a_user, b_user, c_user)

    # Plot the results
    plt.figure(figsize=(10, 6))

    plot_type = input("Enter plot type ('line' or 'scatter'): ").lower()

    if plot_type == 'line':
        plt.plot(time_values, temperature_hardcoded, label='Hardcoded Coefficients', color='blue')
        plt.plot(time_values, temperature_user, label='User Input Coefficients', linestyle='dashed', color='orange')
    elif plot_type == 'scatter':
        plt.scatter(time_values, temperature_hardcoded, label='Hardcoded Coefficients', color='blue')
        plt.scatter(time_values, temperature_user, label='User Input Coefficients', marker='x', color='orange')
    else:
        print("Invalid plot type. Defaulting to line plot.")
        plt.plot(time_values, temperature_hardcoded, label='Hardcoded Coefficients', color='blue')
        plt.plot(time_values, temperature_user, label='User Input Coefficients', linestyle='dashed', color='orange')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation')

    plt.show()

if __name__ == "__main__":
    main()
```
